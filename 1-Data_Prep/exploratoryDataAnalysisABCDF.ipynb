### Inferential Statistics:

* Conduct statistical tests to determine if there are significant differences in student grades based on different categorical features (e.g., gender, region, etc.).

```python
from scipy.stats import ttest_ind

# Example: t-test for difference in final grades between male and female students
male_grades = df[df['GENDER'] == 'M']['PERCENTILE']
female_grades = df[df['GENDER'] == 'F']['PERCENTILE']
t_stat, p_value = ttest_ind(male_grades, female_grades)
print(f"T-Statistic: {t_stat}, P-Value: {p_value}")
```

```python
# Add a new label column to the dataset, called 'GRADE' by binning the 'PERCENTILE' column into the following categories:
# if percentile is less than 60, then GRADE is 'F',
# if percentile is between 60 and 69.9, then GRADE is 'D',
# if percentile is between 70 and 79.9, then GRADE is 'C',
# if percentile is between 80 and 89.9, then GRADE is 'B',
# if percentile is between 90 and 100, then GRADE is 'A',

bins = [0, 60, 70, 80, 90, 101]
labels = ['F', 'D', 'C', 'B', 'A']

df['GRADE'] = pd.cut(df['PERCENTILE'], bins=bins, labels=labels, include_lowest=True)

print(df['GRADE'].value_counts())   
    
```

### Corelation in between a feature and the target variable

A common guideline for determining the strength of correlation between a feature and the target variable (student grades in this case) is to consider the absolute value of the correlation coefficient (Pearson correlation coefficient). The correlation coefficient typically ranges from -1 to 1:

* Positive Correlation: A correlation coefficient close to 1 indicates a strong positive correlation, meaning that as the feature value increases, the target variable (student grades) tends to increase as well.
* Negative Correlation: A correlation coefficient close to -1 indicates a strong negative correlation, meaning that as the feature value increases, the target variable tends to decrease.
* No Correlation: A correlation coefficient close to 0 indicates no linear relationship between the feature and the target variable.

The specific threshold for what constitutes a "strong" correlation can vary depending on the context and domain-specific knowledge. However, as a general guideline:

* A correlation coefficient greater than 0.7 (or less than -0.7) is often considered to indicate a strong correlation.
* A correlation coefficient between 0.5 and 0.7 (or between -0.5 and -0.7) may indicate a moderate correlation.
* A correlation coefficient between 0.3 and 0.5 (or between -0.3 and -0.5) may indicate a weak correlation.

```python
# Import the LabelEncoder from sklearn
from sklearn.preprocessing import LabelEncoder

# List of column names to calculate correlation coefficient against "PERCENTILE"
column_names = ['GENDER', 'EDU_FATHER', 'EDU_MOTHER', 'OCC_FATHER', 'OCC_MOTHER', 'STRATUM', 'SISBEN', 'PEOPLE_HOUSE', 'INTERNET', 'TV', 'COMPUTER', 'WASHING_MCH', 'MIC_OVEN', 'CAR', 'DVD', 'FRESH', 'PHONE', 
                'MOBILE', 'REVENUE', 'JOB', 'SCHOOL_NAME', 'SCHOOL_NAT', 'SCHOOL_TYPE', 'UNIVERSITY', 'ACADEMIC_PROGRAM']

# Encode the categorical columns using LabelEncoder
label_encoder = LabelEncoder()
for column in column_names:
    df[column] = label_encoder.fit_transform(df[column])

# Encode the 'GRADE' columns using LabelEncoder
df['GRADE'] = label_encoder.fit_transform(df['GRADE'])

print(df['GRADE'].value_counts())

# Calculate correlation coefficient for each feature against "GRADE"
correlation = df[column_names].apply(lambda x: x.corr(df['GRADE']))

# Calculate the absolute value of the correlation coefficient
absolute_correlation = correlation.abs()

# Sort the correlation coefficients in descending order
sorted_correlation = absolute_correlation.sort_values(ascending=False)

# Print the sorted correlation coefficients
print(sorted_correlation)

for feature in column_names:
    # Calculate the correlation coefficient
    correlation_coefficient = df[feature].corr(df['GRADE'])

    # Calculate the absolute value of the correlation coefficient
    absolute_correlation_coefficient = abs(correlation_coefficient)

    print(f"Absolute correlation coefficient between '{feature}' and 'GRADE':", round(absolute_correlation_coefficient, 2))
```
